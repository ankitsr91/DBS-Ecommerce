Store Customer Raw File into Delta Table

```python
customers_df = spark.read.format("csv") \
    .option("header", "true") \
        .option("inferSchema","true") \
            .load("/Volumes/dev/bronze/raw_files/Customers.csv")

display(customers_df)

customers_df.write.format("delta") \
    .mode("overwrite") \
        .saveAsTable("dev.bronze.Customers")
```
